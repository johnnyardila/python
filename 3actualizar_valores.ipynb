```python
#lo que haremoa ahora es modificar valores luego de que los extraemos
#primero importamos pandas y nunpy
import pandas as pd
import numpy as np
```

```python
#luego traemos el archivo
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(5)
```

```python
#vamos a fijar el index o Id, para esto usamos el metodo set_index
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores.head(5)
```

```python
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",	"long_name",	"age",	"dob",	"height_cm",	"weight_kg",	"nationality",	"club"]]
df_jugadores.head(5)
```

```python
#vamos a actualizar la estatura de messi pero primero lo vaos a llamar
df_jugadores.loc["L. Messi" , "weight_kg"]
```

```python
#asi se modifica un dato con loc
df_jugadores.loc["L. Messi" , "weight_kg"]=89
```

```python
#aca muestro los datos modificados
df_jugadores.loc["L. Messi" , :]
```

```python
#vamos a modificar la estatura de cristiano ronaldo
df_jugadores.loc["Cristiano Ronaldo" , "height_cm"]=190
```

```python
#vamos a mostrar la iformacion de cristiano
df_jugadores.loc["Cristiano Ronaldo" , :]
```

```python
#vamos actualizar los valores de toda una columna
df_jugadores.loc[:,"height_cm"]=190
```

```python
df_jugadores.loc[:,"height_cm"]
```

```python
#como podria modificar el ultimo registro
df_jugadores.iloc[-1,:]
```

```python
#de esta manera puedo poner datos nulos en el data frame
df_jugadores.iloc[-1, :]=np.nan
```

```python
df_jugadores.iloc[-1,:]
```

```python
#como puedo actualizar multiples valores
#por ejemplo voy actualizar a messi y a cristiano
df_jugadores.iloc[[0,1], 4]=100
```

```python
df_jugadores.head(4)
```

```python
#vamos actualizar valores cuando se cumple con una condicion
#primero definamos que queremos mostrar en una variable
columnas=["age","height_cm","weight","nacionality","club"]
df_jugadores.loc[df_jugadores["height_cm"]>180, columnas]=0
```
