```python
#en este caso reconoceremos el data frame y haremos graficas
#vamos a importar pandas y nunpy
import pandas as pd
import numpy as np
#vamos a traer el archivo de drive
df_poblacion=pd.read_csv("/content/population_total (1).csv", encoding="ISO-8859-1")
df_poblacion.head(5)
```

```python
#lo primero que vamos hacer es eliminar los null o los vahcios, con el metodo dropna
df_poblacion.dropna()
```

```python
#ahora vamos a sobre escribir el data frame
df_poblacion.dropna(inplace=True)
```

```python
#primero pivoteamos para que la visualizacion sea mas amigable
#recordemos que para pivotear tenemos tres argumento index, columnas y valores
df_poblacion.pivot(index="year",columns="country", values="population")
```

```python
#creo un nuevo data frame con la tabla que pivotie
df_pivot_indicadores=df_poblacion.pivot(index="year",columns="country", values="population")
```

```python
#solo traigo los paises que deseo
df_pivot_indicadores[["Colombia", "United States", "India", "China", "Indonesia"]]
```

```python
#guardo en el dataframe el filtro de paises que realice
df_pivot_indicadores=df_pivot_indicadores[["Colombia", "United States", "India", "China", "Indonesia"]]
```

```python
df_pivot_indicadores
```

```python
#verificamos cual es el index
df_pivot_indicadores.index
```

```python
#indentificamos solo el aÃ±o 2020
df_pivot_indicadores.index.isin([2020])
```

```python
#llamamos los datos del aÃ±o 2020
df_pivot_indicadores[df_pivot_indicadores.index.isin([2020])]
```

```python
#guardo en un dataframe nuevo el filtro que hice de solo el 2020
df_pivot_indicadores_2020=df_pivot_indicadores[df_pivot_indicadores.index.isin([2020])]
```

```python
#muestro el nuevo dataframe
df_pivot_indicadores_2020
```

```python
#para transponer uso el metodo T
df_pivot_indicadores_2020.T
```

```python
df_pivot_indicadores_2020=df_pivot_indicadores_2020.T
```

```python
#vamos a graficar con una grafica de barras que se usa con el argumento bar
df_pivot_indicadores_2020.plot(kind="bar", color="orange",
                               xlabel="Pais", ylabel="Poblacion",
                               title="Poblacion 2020")
```
