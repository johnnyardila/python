```python
#tambien podemos eliminar columnas y filas con diferentes metodos
#lo que haremoa ahora es modificar valores luego de que los extraemos
#primero importamos pandas y nunpy
import pandas as pd
import numpy as np
```

```python
#luego traemos el archivo
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(5)
```

```python
#vamos a fijar el index o Id, para esto usamos el metodo set_index
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores.head(5)
```

```python
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",	"long_name",	"age",	"dob",	"height_cm",	"weight_kg",	"nationality",	"club"]]
df_jugadores.head(5)
```

```python
#con el metodo apply voy a calcular la raiz cuadrada de la columna edad
df_jugadores["age"].apply(np.sqrt)
```

```python
df_jugadores["age"].apply(np.square)
```

```python
#para calcular el indice de masa corporal
#creamos la funcion IMC
#IMC calculemos
#el peso dividido por la estatura al cuadrado
#kg/m2
def IMC(row):
  return row["weight_kg"]/((row["height_cm"]/100)**2)
```

```python
#ahora llamamos la funciond
df_jugadores.apply(IMC, axis=1)
```
