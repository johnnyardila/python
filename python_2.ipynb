```python
import pandas as pd #importo pandas y le pongo el sobrenombre pd
import numpy as np #importo numpy y le pongo el sobrenombre np
```

```python
# la primera forma de crear un dataframe es usando el metodo array, y luego guardarlo en una variable
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#para crear el dataframe del arreglo anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2= np.array([[5,10,15],[20,25,30],[35,40,45],[50,55,60]])
```

```python
pd.DataFrame(dataframe_2)
```

para el tercer metodo uso los diccionarios

```python
#primero creamos dos listas, puedo crear cuantas listas quiera entendiendo que las listas son las columnas
ciudadescolombia=["bogota","medellin","cali","barranquilla"]
ciudadesinfo=[11000000,2500000,2200000,1300000]
```
