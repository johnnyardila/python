a diferencia de loc, iloc se usa con los indices de cada elemento esto evita que debamos conocer todos los datos

```python
import pandas as pd
import numpy as np
```

```python
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(5)
```

```python
#vamos a fijar el nombre corto como el index o Id, para esto usamos el metodo set_index
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores.head(5)
```

```python
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",	"long_name",	"age",	"dob",	"height_cm",	"weight_kg",	"nationality",	"club"]]
df_jugadores.head(5)
```

```python
#ahora vamos a usar el iloc
df_jugadores.iloc[1]
```

```python
#para traer los datos es decir una fila y una columna
df_jugadores.iloc[1,4]
```

```python
#para traer el club de neymar
df_jugadores.iloc[2,7]
```

```python
#para traer el nombre completo de todos los jugadores
df_jugadores.iloc[:,1]
```

```python
df_jugadores.iloc[:,6]
```

```python
#como taer los datos de toda la fila de oblak
df_jugadores.iloc[3,:]
```

```python
#como traemos los datos del ultimo jugador
df_jugadores.iloc[-1,:]
```

```python
#ibtener los datos de varias personas
df_jugadores.iloc[[1,2,3] , : ]
```

```python
#ahora vaos a filtrar jugadores que midan mas de 180 y que sean de argentina
df_jugadores.iloc[list((df_jugadores["height_cm"]>180) & (df_jugadores["nationality"]=="Argentina")), : ]
```
