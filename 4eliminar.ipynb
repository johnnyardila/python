```python
#tambien podemos eliminar columnas y filas con diferentes metodos
#lo que haremoa ahora es modificar valores luego de que los extraemos
#primero importamos pandas y nunpy
import pandas as pd
import numpy as np
```

```python
#luego traemos el archivo
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(5)
```

```python
#vamos a fijar el index o Id, para esto usamos el metodo set_index
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores.head(5)
```

```python
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",	"long_name",	"age",	"dob",	"height_cm",	"weight_kg",	"nationality",	"club"]]
df_jugadores.head(5)
```

```python
#vamos a eliminar registros usando el metodo drop
#axis nos sirve para especificar si quiero hacer la accion sobre la columna o sobre la fila
#si es 0 es sobre la fila si es 1 es sobre la columna
df_jugadores.drop("L. Messi", axis=0)
```

```python
#vamos a eliminar teniendo en cuenta el index
df_jugadores.drop(index=["Neymar Jr"])#como es index se entiene que debe eliminar toda la fila
```

```python
#vamos a eliinar dos o mas filas pero las vamos a actualizar
df_jugadores=df_jugadores.drop(["L. Messi","Neymar Jr"], axis=0)
df_jugadores
```

```python
df_jugadores=df_jugadores.drop(["E. Hazard","K. De Bruyne"], axis=0)
df_jugadores
```

```python
#ahora como eliminamos una columna
df_jugadores.drop("player_url", axis=1)#cuando axis es igual a uno, se elimina toda la columna
```

```python
#tambein podemos usar otro metodo para elimianr las columnas sin usar el axis
df_jugadores.drop(columns=["nationality"])
```

```python
#para eliminar la ultima columna por posicion si no la supieramos
df_jugadores.drop(df_jugadores.columns[[-1]],axis=1)
```

```python
df_jugadores.drop(df_jugadores.columns[[6]],axis=1)
```

```python
#ahora como eliminamos mas de una columna
df_jugadores.drop(["long_name", "dob"], axis=1)
```

```python
#para que quede guardados los cambio
df_jugadores=df_jugadores.drop(["long_name", "dob"], axis=1)
df_jugadores
```
