```python
#en este caso reconoceremos el data frame y haremos graficas
#vamos a importar pandas y nunpy
import pandas as pd
import numpy as np
#vamos a traer el archivo de drive
df_poblacion=pd.read_csv("/content/population_total (1).csv", encoding="ISO-8859-1")
df_poblacion.head(5)
```

```python
#lo primero que vamos hacer es eliminar los null o los vahcios, con el metodo dropna
df_poblacion.dropna()
```

```python
#ahora vamos a sobre escribir el data frame
df_poblacion.dropna(inplace=True)
```

```python
#primero pivoteamos para que la visualizacion sea mas amigable
#recordemos que para pivotear tenemos tres argumento index, columnas y valores
df_poblacion.pivot(index="year",columns="country", values="population")
```

```python
#creo un nuevo data frame con la tabla que pivotie
df_poblacion_pivot=df_poblacion.pivot(index="year",columns="country", values="population")
```

```python
df_poblacion_pivot
```

```python
#ahora vamos a llamar algunos paises para no trabajar con todo el dataframe
df_poblacion_pivot[["Colombia", "United States", "India", "China", "Indonesia"]]
```

```python
#sobre escribimos el data frame
df_poblacion_pivot=df_poblacion_pivot[["Colombia", "United States", "India", "China", "Indonesia"]]
df_poblacion_pivot
```

```python
#vamos hacer una grafica de lineas
#usaremos el argumento Kind
#el metodo para graficar es plot
#podemos personalizar la grafica con otros argumentos
#x label me sirve para darle nombre al eje X
#y label me sirve para darle nombre al eje Y
#title me sirve para darle un titulo a la grafica
#figzize me sirve para modificar el tamaño del lienzo
#hay muchos mas
df_poblacion_pivot.plot(kind="line",xlabel="Año",ylabel="Poblacion",
                        title="Poblacion 1955-2020", figsize=(10,5))
```
