```python
#en este caso mezclaremos la funcion lambda con el metodo apply
#la funcion lambda en python es una forma de escribir funciones menos robustas
```

```python
#vamos a importar pandas y nunpy
import pandas as pd
import numpy as np
#vamos a traer el archivo de drive
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(5)
#vamos a fijar el nombre corto como el index o Id, para esto usamos el metodo set_index
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores.head(5)
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",	"long_name",	"age",	"dob",	"height_cm",	"weight_kg",	"nationality",	"club"]]
df_jugadores.head(5)
```

```python
#las funciones normalmente se ven de la siguiente manera
def sumar (a,b):
  x=a+b
  return(x)
```

```python
sumar(29,24)
```

```python
#
def modulo (a,b):
  y=a%b
  return(y)
```

```python
modulo (25,2)
```

```python
#ahora vamos a ver como se hacen las funciones usando lambda
#primero se asigna el nombre
#luego asignamos el valor
#luego introducimos los parametros
#luego dos puntos
#luego el output
sumar_2= lambda a,b:a+b
```

```python
sumar_2(39,39)
```

```python
modulo_2=lambda a,b:a%b
```

```python
modulo_2(25,2)
```

```python
#ahora usemos apply con lambda
#vamos a pasar los centimetro a metros en la informacion de los jugadores
#en este caso si usamos apply no necesitamos poner el nombre de la funcion
df_jugadores["height_cm"].apply(lambda x:x/100)
```

```python
df_jugadores["age"].apply(lambda x:x*12)
```

```python
#si quiero enredarme menos para convertir los cm a metros
df_jugadores["height_cm"]/100
```

```python
#tambien podemos pasar los nombres de los jugadores a mayusculas
df_jugadores["long_name"].apply(lambda x:x.upper())
```

```python
#si quiero hacerlo simple pero con mas memoria seria asi
df_jugadores["long_name"].str.upper()
```

```python
#tambien puedo usar lambda para extraer partes de fechas
#primero debemos garantizar que la columna fecha este en modo date
df_jugadores["dob"]=pd.to_datetime(df_jugadores["dob"])
```

```python
#ahora extraemos los años solo con la funcion lambda
df_jugadores["dob"].apply(lambda x:x.year)
```

```python
#para hallar el nombre del mes se hace de la siguiente manera
df_jugadores["dob"].apply(lambda x:x.month_name())
```
