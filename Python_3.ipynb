```python
#Siempre lo primero que debo hacer es pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde google drive, ya no tenemos que loguearnos de nuevo
df_estudiantes=pd.read_csv('/content/drive/MyDrive/StudentsPerformance (1).csv')
df_estudiantes
```

```python
#Sigmos conla creacion de columnas usando el metodo assign o el metodo insert
#si quiero agregar varias columnas uso assing y tambien si quiero sobre escribir una columna
#creamos notas aleatorias de fisica y de filosofia
filosofia_score=np.random.randint(1, 100, size=1000)
fisica_score=np.random.randint(1, 100, size=1000)
```

```python
#ahora le pondre un indice a cada fila para que sepa
filosofia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy hacer es poner las nuevas columnas a mi dataframe usando assign
df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
```

```python
df_estudiantes=df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
df_estudiantes
```

```python
#Ahora vamos ausar el metodo insert, el metodo insert es diferente por que me permite especificar donde voy a colocar esa columna
#donde voy a colocar esa columna
#al usar el metodo insert debemos tener cuidado por que no me devuelve copias si no el cambio queda hecho de una vez
df_estudiantes.insert(1, 'ed_score', fisica_nota)
```
